```python
x=data[:,:13]
y=data[:,13]
```

```python
##code gradient descent for n features and retrieve the ,i's and c
def gd(x,y):
    params=np.zeros(14)
    lr=0.1
    for i in range(1000):
        support=np.zeros(14)
        for j in range(len(x)):
            xi=x[j,:]
            xi=np.append(xi,1)
            yi=y[j]
            s= (params * xi).sum()
            for k in range(14):
                support[k]=support[k]+ ((-2/len(x)) * (yi-s) * xi[k] )
        ypred=[]
        for j in range(len(x)):
            v=x[j,:]
            v=np.append(v,1)
            ypred.append((v*params).sum())
        ypred=np.array(ypred)
        c=(1/len(x)) * sum([val**2 for val in (y-ypred)])
        print("iteration{},cost{}".format(i+1,c))
        params=params-(lr*support)
    return params
```

```python
params=gd(x,y)
```
